```
#Observando tipos de datos que importa pandas
df_cars.dtypes
```

```
#Cambiando tipo de dato de cada columna para poder operar sobre ellas

df_cars['Carro'] = df_cars['Carro'].astype('str')
df_cars['MPG'] = df_cars['MPG'].astype('float')
df_cars['Cilindros'] = df_cars['Cilindros'].astype('int')
df_cars['Desplazamiento'] = df_cars['Desplazamiento'].astype('int')
df_cars['Caballo_fuerza'] = df_cars['Caballo_fuerza'].astype('int')
df_cars['Peso'] = df_cars['Peso'].astype('int')
df_cars['Aceleracion'] = df_cars['Aceleracion'].astype('float')
df_cars['Modelo'] = df_cars['Modelo'].astype('str')
df_cars['Origen'] = df_cars['Origen'].astype('str')

df_cars.dtypes
```

```
#1.	Encuentra el promedio de "MPG" por país de origen ("Origin").

promedio = df_cars.groupby('Origen').MPG.mean().round(3)
promedio
```

```
#2.	Agrega una columna "Weight_per_Horsepower" que sea el peso dividido por la potencia ("Weight" / "Horsepower"). Además, asegúrense que los valores en esta columna estén redondeados a tres decimales. 

df_cars['Weight_per_Horsepower'] = (df_cars['Peso'] / df_cars['Caballo_fuerza']).round(3)
df_cars.head(5)
```

```
#3.	Encuentra cuántos coches tienen más de 150 caballos de fuerza y una aceleración menor a 10. Además de encontrar estos coches, también se te pide mostrar sus nombres y países de origen. 

coches_seleccionados = df_cars[(df_cars['Caballo_fuerza'] > 150) & (df_cars['Aceleracion'] < 10)]
print(f' Cantidad de coches: {coches_seleccionados.value_counts().sum()}')
coches_seleccionados.head()

```

```
#4.	Encuentra los coches con "MPG" mayor a 25 y "Cylinders" igual a 4 o 6. 
coches_selected = df_cars[(df_cars.MPG > 25) & (df_cars['Cilindros'].isin([4,6]))]
coches_selected.head(10)
```

```
#5.	Encuentra el promedio de "MPG" y "Weight" por país de origen y número de cilindros.

df_average = df_cars.groupby(['Origen','Cilindros']).agg({"MPG":'mean',"Peso":'mean'}).round(3)
df_average.head(10)
```

```
#6.	Encuentra el coche con la aceleración más rápida y el coche con la aceleración más lenta para cada cantidad de cilindros.

#Para poder seleccionar los registros donde se cumplan las condiciones, se necesita encontrar esos indices para luegos seleccionarlos con .loc

coches_seleccionados = df_cars.groupby('Cilindros') \
                         .apply(lambda df: df.loc[[df['Aceleracion'].idxmax(), df['Aceleracion'].idxmin()]]) \
                         .drop_duplicates()

coches_seleccionados.head(10)
```

```
#7.	Encuentra el coche más rápido de cada país de origen ("Origin"). Muestra el nombre del coche, el país de origen y su velocidad máxima.

# NOTA: Para hallar la velocidad maxima, utilicé la siguiente fórmula: Potencia(Caballos de fuerza) * 0.7355

df_cars['velocidad_maxima'] = (df_cars['Caballo_fuerza'] * 0.7355).round(3)
df_carros_idx = df_cars.groupby('Origen').velocidad_maxima.idxmax()
df_cars.loc[df_carros_idx,['Carro','Origen','velocidad_maxima']]
```

```
#8.	Agrupa los coches por país de origen ("Origin") y cuenta cuántos coches hay en cada grupo. 
df_cars.Origen.value_counts()
```

```
#9.	Agrega una nueva columna llamada "Combustible" con un valor predeterminado (por ejemplo, "Gasolina") para todos los coches. 

df_cars['Combustible'] = 'Gasolina'
df_cars.head()
```

```
#10.	Elimina las filas duplicadas basadas en las columnas "Car" y "Model", manteniendo la última aparición.

df_no_duplicados = df_cars.drop_duplicates(subset=['Carro', 'Modelo'], keep='last')
df_no_duplicados.head()
```
